```python
import pandas as pd
df_sub1_esense = pd.read_csv('..\sensor_data\eSense_sbj1.csv', names=['time','Device','1','2','3','4','5','6','Label'], index_col=0)
df_sub1_esense.index = pd.to_datetime(df_sub1_esense.index, unit='ms') + pd.Timedelta('02:00:00')
df_sub1_esense = df_sub1_esense.resample('10L', origin='start').first().interpolate()
#df_sub1_esense.index = df_sub1_esense.index.time
df_sub1_esense.sort_index()
```

```python
df_sub1_ankle_dribbling = pd.read_csv('..\sensor_data\dribbling_sbj1_ankle.csv', index_col=0)
df_sub1_ankle_dribbling.index = pd.date_range(start=df_sub1_ankle_dribbling.index[0], periods = df_sub1_ankle_dribbling.shape[0], freq="10L")

df_sub1_ankle_layup = pd.read_csv('..\sensor_data\layup_sbj1_ankle.csv', index_col=0)
df_sub1_ankle_layup.index = pd.date_range(start=df_sub1_ankle_layup.index[0], periods = df_sub1_ankle_layup.shape[0], freq="10L")

df_sub1_ankle_shooting = pd.read_csv('..\sensor_data\shooting_sbj1_ankle.csv', index_col=0)
df_sub1_ankle_shooting.index = pd.date_range(start=df_sub1_ankle_shooting.index[0], periods = df_sub1_ankle_shooting.shape[0], freq="10L")

df_sub1_ankle_movements = pd.read_csv('..\sensor_data\movements_sbj1_ankle_1.csv', index_col=0)
df_sub1_ankle_movements.index = pd.date_range(start=df_sub1_ankle_movements.index[0], periods = df_sub1_ankle_movements.shape[0], freq="10L")

df_sub1_ankle = pd.concat([df_sub1_ankle_dribbling, df_sub1_ankle_layup, df_sub1_ankle_shooting, df_sub1_ankle_movements])
#df_sub1_ankle.index = pd.date_range(start='2021-04-14 18:22:13.553000', periods = df_sub1_ankle.shape[0], freq="10L")
df_sub1_ankle.sort_index()
```

```python
df_sub1_wrist_dribbling = pd.read_csv('..\sensor_data\dribbling_sbj1_wrist.csv', index_col=0)
df_sub1_wrist_dribbling.index = pd.date_range(start=df_sub1_wrist_dribbling.index[0], periods = df_sub1_wrist_dribbling.shape[0], freq="10L")

df_sub1_wrist_layup = pd.read_csv('..\sensor_data\layup_sbj1_wrist.csv', index_col=0)
df_sub1_wrist_layup.index = pd.date_range(start=df_sub1_wrist_layup.index[0], periods = df_sub1_wrist_layup.shape[0], freq="10L")

df_sub1_wrist_shooting = pd.read_csv('..\sensor_data\shooting_sbj1_wrist.csv', index_col=0)
df_sub1_wrist_shooting.index = pd.date_range(start=df_sub1_wrist_shooting.index[0], periods = df_sub1_wrist_shooting.shape[0], freq="10L")

df_sub1_wrist_movements = pd.read_csv('..\sensor_data\movements_sbj1_wrist.csv', index_col=0)
df_sub1_wrist_movements.index = pd.date_range(start=df_sub1_wrist_movements.index[0], periods = df_sub1_wrist_movements.shape[0], freq="10L")

df_sub1_wrist = pd.concat([df_sub1_wrist_dribbling, df_sub1_wrist_layup, df_sub1_wrist_shooting, df_sub1_wrist_movements])
#df_sub1_wrist.index = pd.date_range(start='2021-04-14 18:32:17.683000', periods = df_sub1_wrist.shape[0], freq="10L")
df_sub1_wrist.sort_index()
```

```python
df_sub1 = pd.concat([df_sub1_esense, df_sub1_ankle, df_sub1_wrist], axis=1, join="inner")
df_sub1
```
